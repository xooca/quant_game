```python
import difflib
def matches(lista, listb):
    list1 = lista.copy()
    list2 = listb.copy()
    while True:
        mbs = difflib.SequenceMatcher(None, list1, list2).get_matching_blocks()
        #print(mbs)
        if len(mbs) == 1: break
        for i, j, n in mbs[::-1]:  
            if n > 0: yield list1[i: i + n]
            del list1[i: i + n]
            del list2[j: j + n]
def get_common_sublist(list1,list2):
    t = list(matches(list1, list2))
    common_list = max(t, key=lambda coll: len(coll))
    for i in range(len(list1)):
        if common_sublist == list1[i:len(common_sublist)+i]:
            start = i
            end = i+len(common_sublist)
    return start,end
```
